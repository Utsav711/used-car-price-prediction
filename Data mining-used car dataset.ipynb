## Data Preprocessing

__for efficient analysis, we'll remove the space between these 2 words company names, for easy data manipulation - Land Rover, Mercedes Benz__

```python
df_cd["Specific Model"].value_counts() #we can simply merge it with the name column
```

```python
df_cd["Transmission"].value_counts()
```

```python
df_cd["Fuel type"].value_counts()
```

```python
df_temp = df_cd.copy()
```

```python
df_temp["Company Name"]=[None]*len(df_temp["name"])
df_temp["year"]=[None]*len(df_temp["name"])
```

```python
for item in range(len(df_temp["name"])):
    df_temp["name"][item]=df_temp["name"][item].split()
    
    df_temp["year"][item]=df_temp["name"][item][0]
    df_temp["Company Name"][item]=df_temp["name"][item][1]
    
    df_temp["name"][item]=df_temp["name"][item][2:]
    df_temp["name"][item]=' '.join(df_temp["name"][item])
```

```python
df_temp["Kms driven"]=df_temp["Kms driven"].str.replace("kms",'')
```

```python
for item in range(len(df_temp["name"])):
    df_temp["name"][item]=df_temp["name"][item].replace("Land Rover","LandRover")
    df_temp["name"][item]=df_temp["name"][item].replace("Mercedes-Benz","MercedesBenz")
```

```python
df_temp['name'] = df_temp[['name', 'Specific Model']].agg(" ".join, axis=1)
```

```python
df_temp.drop("Specific Model",axis=1,inplace=True)
```
